QUES-1-> Describe the main differences between the CSS Flexbox layout model and the CSS Grid layout model. When
would you choose to use one over the other?<br><br>
ANS=>
<BR>
1.> CSS FLEXBOX  : felxbox is designed for one dimensional layouts,meaning it handles items in either a row or a column,but not both at the same time.
                   . it's ideal when oyu want to align and disribute space among items ina container.
                   .flexbox is content-driven the layout is determined by the size and number of items.
                   .we use properties like justify-content,align-items and flex-grow to manage spacing and alignment.

2.> CSS GRID  :  grid is built for two dimensional layouts.it allows you to control both rows and columns simultaneously,making it much more powerful for page layouts.
                . grid id layout -driven=we define a structure first,then place items withi it.
                .it allows for precise control of item postioning,regardless of the order in the html.
                .we cam create complex design like magazines layouts,dashboards and image galleries.

WHEN TO USE WHICH?
ANS=>
.use flexbox when we are dealing with a simple arrangemeny of items in a single row and column.
.use grid when we need a more structures layout involving multiple rows and columns

QUES-2-> Explain the role of the following key properties in the Flexbox layout mode:
1. justify-content
2. align-item/
3. gap
4. flex-direction
5.flex-wrap<br><br>
ANS=>
<br><br>
1->JUSTIFY-CONTENT  : THIS PROPERTY controls hoe flex items are aligned along the main axis.
                .it determines the spacing and alignment of items within the  flex container.
                .commom values=flex-start,center,flex-end,space-between.

2->ALIGN-ITEMS   :  this defines how itmes are alignes along the cross axis.
                   .it sets the alignment for all items inside the container.
                   common values=flex-start,center,flex-end.

3->GAP  :   the gap property controls the spacing between fles items.
           .it sets cosistent spacing between rows and columns.
           .this is cleaner than using margins on individual items.

4->FLEX-DIRECTION  :  defines the direction in which flex items are palced in the container.
                    .options=row,column.row-reverse,column-reverse.
                    .this property sets what is considered the main axis.

5->FLEX-WRAP  :  controls whether flex-items stay ina single line por wrap onto multiple lines.
                .options=nowrap,wrap,wrap-reverse.
                .useful when we have many items or small screen sizes.
